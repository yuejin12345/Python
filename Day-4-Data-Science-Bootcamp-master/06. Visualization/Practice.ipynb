```python
# Q4: Make a column datatime
flights.head()
```

```python
flights['time'] = flights['year'].astype(str) + '/' + flights['month'].astype(str) +'/' +  flights['day'].astype(str)
```

```python
flights.head()
```

```python
flights['time1'] = pd.to_datetime(flights['time'], format='%Y/%m/%d')
```

```python
flights.dtypes
```
